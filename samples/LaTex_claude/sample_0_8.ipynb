# The Physics of Musical Harmony
**Category: Physics**

## 1. Introduction

Why does a perfect fifth sound pleasing while a minor second grates? The answer lies in physics, mathematics, and neuroscience. Musical harmony is governed by the overtone series, interference patterns between sound waves, and the way the auditory cortex processes periodic stimuli. The same mathematics that describes vibrating strings also explains why Western scales are tuned the way they are — and why no tuning system is ever quite perfect.

## 2. Vibrating Strings and the Overtone Series

### 2.1 Standing Waves on a String

A string of length $L$, linear mass density $\mu$, and tension $T$ supports standing waves at frequencies
$$
f_n = \frac{n}{2L}\sqrt{\frac{T}{\mu}}, \qquad n = 1, 2, 3, \ldots
$$
The fundamental $f_1$ is the perceived pitch. The overtones $f_2, f_3, \ldots$ (harmonics) give the instrument its timbre. A plucked string excites all harmonics simultaneously; their relative amplitudes depend on the plucking position.

### 2.2 The Harmonic Series and Consonance

Two tones with fundamental frequencies $f_a$ and $f_b$ are consonant when their overtone series share many coincident partials. Coincidence occurs when $f_a / f_b$ is a simple rational number:

| Interval | Frequency ratio | Consonance |
|---|---|---|
| Unison | $1:1$ | maximal |
| Octave | $2:1$ | maximal |
| Perfect fifth | $3:2$ | very high |
| Perfect fourth | $4:3$ | high |
| Major third | $5:4$ | moderate |
| Minor second | $16:15$ | dissonant |

The **critical bandwidth** of the cochlea ($\sim 1/3$ octave around any frequency) sets the window within which two simultaneous partials produce beating and roughness rather than fusing into consonance.

## 3. Temperament and Tuning

### 3.1 Pythagorean Tuning and the Comma

Stacking 12 pure fifths ($3:2$) should return to the starting pitch after 7 octaves. But
$$
\left(\frac{3}{2}\right)^{12} = \frac{3^{12}}{2^{12}} = \frac{531441}{4096} \approx 129.746 \neq 2^7 = 128
$$
The discrepancy, called the **Pythagorean comma**, is $\approx 23.46$ cents. No tuning system based on pure intervals can close the circle of fifths exactly — the mathematics is fundamentally incompatible.

### 3.2 Equal Temperament

The modern solution is **12-tone equal temperament** (12-TET): divide the octave into 12 equal semitones, each a frequency ratio of $2^{1/12}$. The fifth becomes $2^{7/12} \approx 1.4983$ instead of $1.5000$ — off by only $1.96$ cents. In cents (hundredths of a semitone, $1200$ per octave):
$$
c = 1200 \log_2\!\left(\frac{f}{f_0}\right)
$$

## 4. Numerical Exploration

### 4.1 Beating Between Mistuned Intervals

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 44100
t = np.linspace(0, 1.0, fs, endpoint=False)

f1 = 440.0  # A4
f2_just  = f1 * 3/2          # pure fifth: 660 Hz
f2_et    = f1 * 2**(7/12)    # equal-tempered fifth
f2_minor2 = f1 * 2**(1/12)   # minor second

wave_just   = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2_just*t)
wave_et     = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2_et*t)
wave_minor2 = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2_minor2*t)

fig, axes = plt.subplots(3, 1, figsize=(10, 7), sharex=True)
t_ms = t * 1000
zoom = t_ms < 50

axes[0].plot(t_ms[zoom], wave_just[zoom], linewidth=1)
axes[0].set_title(f'Just fifth: 440 Hz + {f2_just:.1f} Hz (ratio 3:2)')
axes[1].plot(t_ms[zoom], wave_et[zoom], linewidth=1, color='orange')
axes[1].set_title(f'Equal-tempered fifth: 440 Hz + {f2_et:.2f} Hz (ratio $2^{{7/12}}$)')
axes[2].plot(t_ms[zoom], wave_minor2[zoom], linewidth=1, color='red')
axes[2].set_title(f'Minor second: 440 Hz + {f2_minor2:.2f} Hz (ratio $2^{{1/12}}$)')

for ax in axes:
    ax.set_ylabel('Amplitude')
axes[2].set_xlabel('Time (ms)')
plt.tight_layout()
plt.show()
```

### 4.2 The Circle of Fifths — Deviation from Pure Intervals

```python
notes = ['C','G','D','A','E','B','F#','C#','G#','D#','A#','F']
pure_fifth_cents = 702.0
et_fifth_cents   = 700.0
pythag_stack = [(i * pure_fifth_cents) % 1200 for i in range(12)]
et_stack     = [(i * et_fifth_cents)   % 1200 for i in range(12)]
diff         = [p - e for p, e in zip(pythag_stack, et_stack)]

fig, ax = plt.subplots(figsize=(9, 4))
x = np.arange(12)
ax.bar(x, diff, color='steelblue', alpha=0.8)
ax.set_xticks(x)
ax.set_xticklabels(notes)
ax.axhline(0, color='black', linewidth=0.8)
ax.set_ylabel('Pythagorean − Equal Temperament (cents)')
ax.set_title('Tuning Deviation: Pythagorean vs. 12-TET Across the Circle of Fifths')
plt.tight_layout()
plt.show()
```
