# Convergence of Fourier Series
**Category: Analysis**

## 1. Introduction

The Fourier series is a decomposition of a periodic function $f : \mathbb{R} \to \mathbb{C}$ into a sum of sinusoids. It is one of the most powerfull tools in classical analysis.

## 2. Theory

### 2.1 Definition

Let $f \in L^2([-\pi, \pi])$. The Fourier series of $f$ is
$$
f(x) \sim \sum_{n=-\infty}^{\infty} \hat{f}(n)\, e^{inx}
$$
where the Fourier coefficients are given by
$$
\hat{f}(n) = \frac{1}{2\pi} \int_{-\pi}^{\pi} f(x)\, e^{-inx}\, dx, \quad n \in \mathbb{Z}
$$

### 2.2 Convergence

#### 2.2.1 $L^2$ Convergence

For any $f \in L^2([-\pi,\pi])$, the partial sums $S_N f(x) = \sum_{|n| \leq N} \hat{f}(n) e^{inx}$ satisfy
$$
\lim_{N \to \infty} \| S_N f - f \|_{L^2} = 0
$$
and Parseval's identity holds:
$$
\| f \|_{L^2}^2 = 2\pi \sum_{n=-\infty}^{\infty} |\hat{f}(n)|^2
$$

#### 2.2.2 Pointwise Convergence

If $f$ is continuously differentiable at $x_0$, then $S_N f(x_0) \to f(x_0)$ as $N \to \infty$. More precisely, the Dirichlet kernel $D_N$ satisfies
$$
S_N f(x) = (f * D_N)(x) = \frac{1}{2\pi} \int_{-\pi}^{\pi} f(x - t)\, D_N(t)\, dt
$$
where
$$
D_N(t) = \sum_{n=-N}^{N} e^{int} = \frac{\sin\bigl((N + \tfrac{1}{2})t\bigr)}{\sin(t/2)}
$$

##### 2.2.2.1 Gibbs Phenomenon

At a jump discontinuity of $f$, the partial sums overshoot by approximately $9\%$ regardless of $N$. Precisely, if $f$ has a jump of size $\delta = f(x_0^+) - f(x_0^-)$, then
$$
\limsup_{N\to\infty} S_N f(x) \approx f(x_0^+) + 0.0895\, \delta
$$

### 3.2 Decay of Coefficients

For the square wave $f(x) = \mathrm{sgn}(x)$, the Fourier coefficients satisfy $\hat{f}(n) = 0$ for even $n$, and
$$
\hat{f}(n) = \frac{2}{i\pi n}, \quad n \text{ odd}
$$
so $|\hat{f}(n)| = O(n^{-1})$, consistent with the $C^0 \setminus C^1$ regularity of $f$. In general, $k$ derivatives in $L^2$ implies $|\hat{f}(n)| = O(n^{-k})$.
