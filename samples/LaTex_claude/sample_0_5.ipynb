## 1. Introduction

Claude Shannon's 1948 paper "A Mathematical Theory of Communication" founded information theory by giving a precise, quantitative definition of **information** and **uncertainty**. The resulting framework underpins everything from data compression and error-correcting codes to machine learning, cryptography, and neuroscience.

## 2. Shannon Entropy

### 2.1 Definition

Let $X$ be a discrete random variable taking values in a finite alphabet $\mathcal{X}$ with probability mass function $p(x) = P(X = x)$. The **Shannon entropy** of $X$ is
$$
H(X) = -\sum_{x \in \mathcal{X}} p(x) \log_2 p(x)
$$
measured in **bits** (using $\log_2$). The convention $0 \log_2 0 = 0$ applies. $H(X)$ quantifies the average number of bits needed to describe the outcome of $X$, or equivalently, our uncertainty about $X$ before observing it.

### 2.2 Properties

- **Non-negativity**: $H(X) \geq 0$, with equality iff $X$ is deterministic.
- **Maximum entropy**: $H(X) \leq \log_2 |\mathcal{X}|$, with equality iff $X$ is uniform.
- **Concavity**: $H$ is a concave function of $p$.
- **Chain rule**: $H(X, Y) = H(X) + H(Y \mid X)$, where the conditional entropy is
$$
H(Y \mid X) = -\sum_{x, y} p(x, y) \log_2 p(y \mid x)
$$

## 3. Mutual Information and KL Divergence

### 3.1 Mutual Information

The **mutual information** $I(X; Y)$ measures the reduction in uncertainty about $X$ after observing $Y$:
$$
I(X; Y) = H(X) - H(X \mid Y) = \sum_{x,y} p(x,y) \log_2 \frac{p(x,y)}{p(x)\,p(y)}
$$
$I(X;Y) = 0$ iff $X$ and $Y$ are independent. It is symmetric: $I(X;Y) = I(Y;X)$.

### 3.2 KL Divergence

The **Kullback–Leibler divergence** from distribution $q$ to distribution $p$ is
$$
D_{\mathrm{KL}}(p \| q) = \sum_{x} p(x) \log_2 \frac{p(x)}{q(x)}
$$
It satisfies $D_{\mathrm{KL}}(p \| q) \geq 0$ with equality iff $p = q$ (Gibbs' inequality), but it is **not** symmetric and not a metric. Mutual information can be written as $I(X;Y) = D_{\mathrm{KL}}(p_{XY} \| p_X \otimes p_Y)$.

## 4. Source Coding

### 4.1 Shannon's Source Coding Theorem

For an i.i.d. source with entropy $H(X)$ bits per symbol, the minimum average code length $\bar{L}$ of any uniquely decodable code satisfies
$$
H(X) \leq \bar{L} < H(X) + 1
$$
This bound is achieved asymptotically by Huffman coding (for a single symbol) and arithmetic coding (for long sequences). No lossless compression can do better than $H(X)$ bits per source symbol.

## 5. Numerical Experiment

### 5.1 Entropy of English Text

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

def entropy_bits(counts):
    total = sum(counts.values())
    probs = np.array([v / total for v in counts.values()])
    probs = probs[probs > 0]
    return -np.sum(probs * np.log2(probs))

# Approximate letter frequencies for English (relative)
english_freq = {
    'e': 12.70, 't': 9.06, 'a': 8.17, 'o': 7.51, 'i': 6.97, 'n': 6.75,
    's': 6.33, 'h': 6.09, 'r': 5.99, 'd': 4.25, 'l': 4.03, 'c': 2.78,
    'u': 2.76, 'm': 2.41, 'w': 2.36, 'f': 2.23, 'g': 2.02, 'y': 1.97,
    'p': 1.93, 'b': 1.49, 'v': 0.98, 'k': 0.77, 'j': 0.15, 'x': 0.15,
    'q': 0.10, 'z': 0.07
}

H_english = entropy_bits(english_freq)
H_uniform = np.log2(26)

print(f'Entropy of English letter distribution: {H_english:.4f} bits')
print(f'Entropy of uniform distribution over 26 letters: {H_uniform:.4f} bits')
print(f'Redundancy: {(1 - H_english / H_uniform) * 100:.1f}%')
```

```python
# Plot entropy of a binary source as a function of p
p_vals = np.linspace(1e-6, 1 - 1e-6, 1000)
H_binary = -(p_vals * np.log2(p_vals) + (1 - p_vals) * np.log2(1 - p_vals))

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].plot(p_vals, H_binary, linewidth=2, color='steelblue')
axes[0].axvline(0.5, color='gray', linestyle='--', label='$p=0.5$, $H=1$ bit')
axes[0].set_xlabel('$p$')
axes[0].set_ylabel('$H(p)$ (bits)')
axes[0].set_title('Binary Entropy Function')
axes[0].legend()

letters = list(english_freq.keys())
freqs = list(english_freq.values())
axes[1].bar(letters, freqs, color='steelblue', alpha=0.8)
axes[1].set_xlabel('Letter')
axes[1].set_ylabel('Frequency (%)')
axes[1].set_title(f'English Letter Frequencies  ($H = {H_english:.2f}$ bits)')

plt.tight_layout()
plt.show()
```
