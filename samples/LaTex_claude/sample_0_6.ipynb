# Predator–Prey Dynamics: The Lotka–Volterra System
**Category: Biology**

## 1. Introduction

The Lotka–Volterra equations, formulated independently by Alfred Lotka (1925) and Vito Volterra (1926), are the canonical mathematical model of predator–prey interaction. They describe periodic oscillations in population sizes — a pattern observed in real ecosystems such as the Canadian lynx and snowshoe hare cycles recorded by the Hudson's Bay Company over more than a century.

## 2. The Model

### 2.1 Equations of Motion

Let $x(t)$ be the prey population and $y(t)$ the predator population. The system is
$$
\frac{dx}{dt} = \alpha x - \beta x y
$$
$$
\frac{dy}{dt} = \delta x y - \gamma y
$$
where:
- $\alpha > 0$ — intrinsic prey birth rate
- $\beta > 0$ — predation rate per predator–prey encounter
- $\delta > 0$ — predator reproduction rate per prey consumed
- $\gamma > 0$ — intrinsic predator death rate

### 2.2 Fixed Points

Setting $\dot{x} = \dot{y} = 0$ yields two fixed points:
- **Trivial**: $(x^*, y^*) = (0, 0)$ — extinction of both species
- **Coexistence**: $(x^*, y^*) = \left(\dfrac{\gamma}{\delta},\ \dfrac{\alpha}{\beta}\right)$

Linearisation around the coexistence fixed point yields purely imaginary eigenvalues $\pm i\sqrt{\alpha\gamma}$, confirming that the system exhibits **neutral oscillations** — closed orbits in phase space whose amplitude depends on initial conditions.

### 2.3 Conservation Law

The system possesses a conserved quantity (first integral):
$$
V(x, y) = \delta x - \gamma \ln x + \beta y - \alpha \ln y = \text{const.}
$$
This can be verified by computing $\dot{V} = 0$ along trajectories. Each orbit is a level curve of $V$, confirming the closed-orbit structure.

## 3. Numerical Simulation

### 3.1 Time Series and Phase Portrait

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

alpha, beta, delta, gamma = 1.0, 0.1, 0.075, 1.5

def lotka_volterra(t, y):
    x, p = y
    return [alpha*x - beta*x*p, delta*x*p - gamma*p]

# Multiple initial conditions
ics = [(10, 5), (15, 5), (20, 5), (10, 10)]
t_span = (0, 40)
t_eval = np.linspace(*t_span, 2000)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Time series for one trajectory
sol = solve_ivp(lotka_volterra, t_span, ics[0], t_eval=t_eval, method='RK45', rtol=1e-8)
axes[0].plot(sol.t, sol.y[0], label='Prey $x(t)$', linewidth=2)
axes[0].plot(sol.t, sol.y[1], label='Predator $y(t)$', linewidth=2, linestyle='--')
axes[0].axhline(gamma/delta, color='gray', linestyle=':', label=f'$x^* = {gamma/delta:.1f}$')
axes[0].axhline(alpha/beta, color='orange', linestyle=':', label=f'$y^* = {alpha/beta:.1f}$')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Population')
axes[0].set_title('Lotka–Volterra: Time Series')
axes[0].legend(fontsize=8)

# Phase portrait
for x0, y0 in ics:
    sol = solve_ivp(lotka_volterra, t_span, [x0, y0], t_eval=t_eval, method='RK45', rtol=1e-8)
    axes[1].plot(sol.y[0], sol.y[1], linewidth=1.5)
axes[1].plot(gamma/delta, alpha/beta, 'ko', markersize=7, label='Coexistence fixed point')
axes[1].set_xlabel('Prey $x$')
axes[1].set_ylabel('Predator $y$')
axes[1].set_title('Phase Portrait')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### 3.2 Verification of the Conservation Law

```python
sol = solve_ivp(lotka_volterra, (0, 200), ics[0], t_eval=np.linspace(0, 200, 5000), rtol=1e-10, atol=1e-12)
x_t, y_t = sol.y
V = delta*x_t - gamma*np.log(x_t) + beta*y_t - alpha*np.log(y_t)

fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(sol.t, V - V[0], linewidth=1.5)
ax.set_xlabel('Time')
ax.set_ylabel('$V(t) - V(0)$')
ax.set_title('Conservation Law Drift (numerical error only)')
plt.tight_layout()
plt.show()
```
