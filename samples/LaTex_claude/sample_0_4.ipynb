# The Chemistry of Coffee
**Category: Chemistry**

## 1. Introduction

A cup of coffee contains over 1,000 distinct chemical compounds. Its flavour, aroma, and stimulant effect arise from a precise interplay of alkaloids, acids, sugars, lipids, and volatile organic compounds — all shaped by roasting temperature, grind size, water chemistry, and extraction time. Understanding the chemistry behind coffee is, in a sense, understanding a small-scale chemical reactor that fits in your kitchen.

## 2. Key Compounds

### 2.1 Caffeine

Caffeine ($\text{C}_8\text{H}_{10}\text{N}_4\text{O}_2$, molar mass $194.19\,\text{g/mol}$) is a methylxanthine alkaloid. It competitively inhibits adenosine receptors (primarily $\text{A}_1$ and $\text{A}_{2A}$), blocking the accumulation of the sleep-promoting nucleoside adenosine. Its half-life in the human body is approximately $5$–$6$ hours, though this varies with liver enzyme activity (CYP1A2 polymorphisms).

A standard espresso ($\sim 30\,\text{mL}$) contains roughly $60$–$80\,\text{mg}$ of caffeine; a drip-brew cup ($\sim 240\,\text{mL}$) contains $90$–$150\,\text{mg}$.

### 2.2 Chlorogenic Acids

Chlorogenic acids (CGAs) are esters of hydroxycinnamic acids (primarily caffeic acid) with quinic acid. They account for $6$–$12\%$ of the dry weight of green coffee and are the primary source of the perceived acidity and bitterness after roasting. During roasting, CGAs hydrolyse and isomerise into quinolactones, which are among the most bitter compounds in coffee. The degree of CGA degradation is a direct proxy for roast degree.

### 2.3 Maillard Reaction and Caramelisation

The brown colour and toasted aroma of roasted coffee arise from two non-enzymatic browning reactions:

- **Maillard reaction**: between reducing sugars and amino acids, producing hundreds of heterocyclic compounds (pyrazines, furans, thiophenes) responsible for roasted, nutty, and caramel-like notes. Begins at $\sim 140^\circ\text{C}$.
- **Caramelisation**: thermal degradation of sucrose and other sugars at $\sim 170^\circ\text{C}$, yielding caramel flavour compounds and CO$_2$.

Both reactions depend critically on temperature and time, making roast profiling as much chemistry as craft.

## 3. Extraction Kinetics

### 3.1 The Extraction Yield

The **extraction yield** (EY) is the fraction of soluble mass extracted from the grounds:
$$
\text{EY} = \frac{m_{\text{dissolved}}}{m_{\text{grounds}}}
$$
The Specialty Coffee Association defines the ideal EY window as $18$–$22\%$. Under-extraction ($<18\%$) gives sour, grassy flavours; over-extraction ($>22\%$) gives harsh bitterness. EY depends on grind size $d$, water temperature $T$, contact time $t$, and pressure $P$.

### 3.2 A Simple Diffusion Model

Soluble compounds diffuse out of a spherical coffee particle of radius $R$ into the surrounding water. Assuming first-order kinetics with rate constant $k$, the concentration of solubles in solution $C(t)$ evolves as
$$
\frac{dC}{dt} = k\bigl(C_{\text{eq}} - C(t)\bigr)
$$
with solution $C(t) = C_{\text{eq}}\left(1 - e^{-kt}\right)$. The rate $k$ increases with temperature (Arrhenius) and decreases with particle size ($k \propto R^{-2}$ from Fick's law), which is why finer grinds extract faster.

## 4. Numerical Experiment

### 4.1 Extraction Curves

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 300, 500)  # seconds

# First-order extraction model C(t) = C_eq * (1 - exp(-k*t))
# k scales inversely with grind size^2 and with temperature via Arrhenius
scenarios = [
    ('Coarse grind, 90°C',  0.010),
    ('Medium grind, 90°C', 0.018),
    ('Fine grind, 90°C',   0.030),
    ('Fine grind, 95°C',   0.040),
]

C_eq = 22.0  # % max extractable

fig, ax = plt.subplots(figsize=(8, 4))
for label, k in scenarios:
    C = C_eq * (1 - np.exp(-k * t))
    ax.plot(t, C, label=label, linewidth=2)

ax.axhspan(18, 22, alpha=0.15, color='green', label='SCA ideal range (18–22%)')
ax.set_xlabel('Extraction time (s)')
ax.set_ylabel('Extraction yield (%)')
ax.set_title('Coffee Extraction Curves (First-Order Diffusion Model)')
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

### 4.2 Caffeine Pharmacokinetics

```python
# One-compartment pharmacokinetic model: oral caffeine absorption and elimination
# dC/dt = F * D * k_a * exp(-k_a * t) / V - k_e * C
# Closed-form solution:
# C(t) = (F * D * k_a) / (V * (k_a - k_e)) * (exp(-k_e*t) - exp(-k_a*t))

F = 1.0        # bioavailability (nearly 100% for caffeine)
D = 80.0       # dose (mg)
k_a = 0.693 / 0.75   # absorption half-life ~45 min
k_e = 0.693 / 5.5    # elimination half-life ~5.5 h
V = 40.0       # apparent volume of distribution (L)

t_h = np.linspace(0, 24, 500)  # hours
C_plasma = (F * D * k_a) / (V * (k_a - k_e)) * (np.exp(-k_e * t_h) - np.exp(-k_a * t_h))

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t_h, C_plasma, linewidth=2, color='saddlebrown')
ax.axhline(1.5, color='gray', linestyle='--', label='typical alertness threshold (~1.5 mg/L)')
ax.set_xlabel('Time after dose (h)')
ax.set_ylabel('Plasma concentration (mg/L)')
ax.set_title('Caffeine Pharmacokinetics (80 mg oral dose)')
ax.legend()
plt.tight_layout()
plt.show()
```
