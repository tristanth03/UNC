# Black Holes and the Schwarzschild Metric
**Category: Physics**

## 1. Introduction

A black hole is a region of spacetime where gravity is so strong that nothing — not even light — can escape once it crosses the **event horizon**. The first exact solution to Einstein's field equations describing such an object was found by Karl Schwarzschild in 1916, just months after Einstein published the general theory of relativity.

## 2. The Schwarzschild Solution

### 2.1 The Metric

For a static, spherically symmetric, uncharged mass $M$ in vacuum, the spacetime interval is
$$
ds^2 = -\left(1 - \frac{r_s}{r}\right)c^2\,dt^2 + \left(1 - \frac{r_s}{r}\right)^{-1}dr^2 + r^2 d\Omega^2
$$
where $d\Omega^2 = d\theta^2 + \sin^2\theta\, d\phi^2$ is the solid angle element, and the **Schwarzschild radius** is
$$
r_s = \frac{2GM}{c^2}
$$
At $r = r_s$ the metric component $g_{tt}$ vanishes and $g_{rr}$ diverges — this is the event horizon. The singularity at $r = 0$ is a genuine curvature singularity.

### 2.2 Gravitational Time Dilation

A clock at rest at coordinate radius $r$ ticks slower than a clock at infinity. The ratio of proper time $d\tau$ to coordinate time $dt$ is
$$
\frac{d\tau}{dt} = \sqrt{1 - \frac{r_s}{r}}
$$
As $r \to r_s^+$, time dilation becomes infinite — a distant observer watches infalling matter freeze at the horizon, redshifting to invisibility.

### 2.3 Photon Sphere and ISCO

Two special radii govern the structure around a Schwarzschild black hole:

| Feature | Radius | Description |
|---|---|---|
| Event horizon | $r_s = 2GM/c^2$ | point of no return |
| Photon sphere | $r_{\gamma} = \tfrac{3}{2}r_s$ | unstable circular photon orbit |
| ISCO | $r_{\text{ISCO}} = 3r_s$ | innermost stable circular orbit for massive particles |

The ISCO sets the inner edge of accretion discs and is directly observable in X-ray binaries.

## 3. Hawking Radiation

### 3.1 Thermal Emission

Quantum field theory in curved spacetime predicts that black holes radiate thermally at the **Hawking temperature**
$$
T_H = \frac{\hbar c^3}{8\pi G M k_B}
$$
This is inversely proportional to mass: stellar-mass black holes ($M \sim 10 M_\odot$) have $T_H \sim 10^{-8}\,\text{K}$, far below the CMB temperature, and evaporation timescales far exceed the age of the universe.

### 3.2 Evaporation Timescale

By Stefan–Boltzmann radiation and energy conservation, the evaporation time scales as
$$
t_{\text{evap}} \sim \frac{5120\, \pi\, G^2 M^3}{\hbar\, c^4}
$$
For a primordial black hole to be evaporating today ($t_{\text{evap}} \approx 13.8\,\text{Gyr}$), its initial mass must have been $M \approx 5 \times 10^{11}\,\text{kg}$ — about the mass of a mountain.

## 4. Numerical Exploration

### 4.1 Schwarzschild Radii of Familiar Objects

```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.674e-11
c = 2.998e8
hbar = 1.055e-34
k_B = 1.381e-23

objects = {
    'Earth':      5.972e24,
    'Sun':        1.989e30,
    'Stellar BH (10 Msun)': 10 * 1.989e30,
    'Sgr A* (4e6 Msun)':    4e6 * 1.989e30,
    'M87* (6.5e9 Msun)':    6.5e9 * 1.989e30,
}

print(f'{'Object':<30} {'Mass (kg)':>15} {'r_s (m)':>15} {'T_H (K)':>15}')
print('-' * 80)
for name, M in objects.items():
    rs = 2 * G * M / c**2
    TH = hbar * c**3 / (8 * np.pi * G * M * k_B)
    print(f'{name:<30} {M:>15.3e} {rs:>15.3e} {TH:>15.3e}')
```

### 4.2 Time Dilation Profile

```python
r_over_rs = np.linspace(1.01, 10, 1000)
dilation = np.sqrt(1 - 1 / r_over_rs)

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(r_over_rs, dilation, linewidth=2)
ax.axvline(1.5, color='gray', linestyle='--', label='photon sphere ($1.5\,r_s$)')
ax.axvline(3.0, color='orange', linestyle='--', label='ISCO ($3\,r_s$)')
ax.set_xlabel('$r / r_s$')
ax.set_ylabel('$d\\tau / dt$')
ax.set_title('Gravitational Time Dilation near a Schwarzschild Black Hole')
ax.legend()
plt.tight_layout()
plt.show()
```
